```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Using iris dataset
from ucimlrepo import fetch_ucirepo

# fetch dataset
iris = fetch_ucirepo(id=53)

# data (as pandas dataframes)
X = iris.data.features
y = iris.data.targets

# metadata
print(iris.metadata)

# variable information
print(iris.variables)
```

```python
# Initialize variables
num_runs = 100
nb_accuracies = []
knn_accuracies = []

# Perform multiple runs
for _ in range(num_runs):
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=None)

    # Naive Bayes Classifier
    nb_model = GaussianNB()
    nb_model.fit(X_train, y_train)
    nb_predictions = nb_model.predict(X_test)
    nb_accuracy = accuracy_score(y_test, nb_predictions)
    nb_accuracies.append(nb_accuracy)

    # K-Nearest Neighbors Classifier
    knn_model = KNeighborsClassifier(n_neighbors=5)
    knn_model.fit(X_train, y_train)
    knn_predictions = knn_model.predict(X_test)
    knn_accuracy = accuracy_score(y_test, knn_predictions)
    knn_accuracies.append(knn_accuracy)
```

```python
#accuracy
avg_nb_accuracy = np.mean(nb_accuracies)
avg_knn_accuracy = np.mean(knn_accuracies)

# avg
print(f"Average accuracy of Naive Bayes: {avg_nb_accuracy:.2f}")
print(f"Average accuracy of KNN: {avg_knn_accuracy:.2f}")
```

```python
plt.figure(figsize=(14, 7))
plt.plot(range(num_runs), nb_accuracies, label='Naive Bayes', marker='o', color='black', linestyle='dashed')
plt.plot(range(num_runs), knn_accuracies, label='KNN', marker='x', color='gray', linestyle='solid')
plt.xlabel('Run')
plt.ylabel('Accuracy')
plt.title('Accuracies of Naive Bayes and KNN over Multiple Runs')
plt.legend()
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
# Initialize variables
num_runs = 100
dt_accuracies = []
knn_accuracies = []

# Perform multiple runs
for _ in range(num_runs):
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=None)

    # Decision Tree Classifier
    dt_model = DecisionTreeClassifier()
    dt_model.fit(X_train, y_train)
    dt_predictions = dt_model.predict(X_test)
    dt_accuracy = accuracy_score(y_test, dt_predictions)
    dt_accuracies.append(dt_accuracy)

    # K-Nearest Neighbors Classifier
    knn_model = KNeighborsClassifier(n_neighbors=5)
    knn_model.fit(X_train, y_train)
    knn_predictions = knn_model.predict(X_test)
    knn_accuracy = accuracy_score(y_test, knn_predictions)
    knn_accuracies.append(knn_accuracy)
```

```python
avg_dt_accuracy = np.mean(dt_accuracies)
avg_knn_accuracy = np.mean(knn_accuracies)

# Print avg accuracy
print(f"Average accuracy of Decision Tree: {avg_dt_accuracy:.2f}")
print(f"Average accuracy of KNN: {avg_knn_accuracy:.2f}")

```

```python
# Plot accuracy
plt.figure(figsize=(14, 7))
plt.plot(range(num_runs), dt_accuracies, label='Decision Tree', marker='o', color='black', linestyle='dashed')
plt.plot(range(num_runs), knn_accuracies, label='KNN', marker='x', color='gray', linestyle='solid')
plt.xlabel('Run')
plt.ylabel('Accuracy')
plt.title('Accuracies of Decision Tree and KNN over Multiple Runs')
plt.legend()
plt.show()
```
