```python
import pandas as pd
import numpy as np
df=pd.read_csv('H_DDataset.csv')
df
```

```python
correlation = df['max heart rate'].corr(df['resting bp s'])

print(correlation)
```

```python
CT = pd.crosstab(df['chest pain type'], df['fasting blood sugar'],margins=True)
CT
observed = CT.values
observed
```

```python
expected = np.zeros(observed.shape)
row_sums = observed.sum(axis=1)
col_sums = observed.sum(axis=0)
total = observed.sum()

for i in range(len(row_sums)):
    for j in range(len(col_sums)):
        expected[i, j] = (row_sums[i] * col_sums[j]) / total

expected
```

```python
from scipy.stats import chi2
chi_square_stat = ((observed - expected) ** 2 / expected).sum()

# Degrees of freedom
dof = (observed.shape[0] - 1) * (observed.shape[1] - 1)

# p-value
p_value = 1 - chi2.cdf(chi_square_stat, dof)

chi_square_stat
```
