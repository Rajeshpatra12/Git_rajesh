```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv("C:\\Users\\rhpat\\OneDrive\\Documents\\dataset.csv")
df
```

```python
print(df.head())
disease_distribution = df['target'].value_counts()
print(disease_distribution)
```

```python
#1.Plot the Distribution of disease
from matplotlib import pyplot as plt
plt.figure(figsize=(10,10)) # set the figure size
plt.pie(disease_distribution, labels=['No Disease', 'Disease'], autopct='%1.1f%%' , startangle=90, colors=['#FFFF00','#A020F0'])
plt.title('Distribution of Disease')
plt.ylabel('') # Hide the y-axis label
plt.show()
```

```python
#2. Plot attribute statistics
plt.figure(figsize=(14,10))
df.hist(bins=20, figsize=(14, 10))
plt.suptitle('Attribute Statistics')
plt.show()
```
