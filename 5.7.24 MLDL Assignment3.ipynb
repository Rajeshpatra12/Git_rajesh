```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt


from ucimlrepo import fetch_ucirepo

# dataset
iris = fetch_ucirepo(id=53)

X = iris.data.features
y = iris.data.targets

print(iris.metadata)

print(iris.variables)

```

```python
# Applying PCA without norma
pca = PCA(n_components=2)
principal_components_no_norm = pca.fit_transform(X)
principal_components_no_norm
```

```python
# Applying PCA with norma
scaler = StandardScaler()
df_normalized = scaler.fit_transform(X)
pca_normalized = PCA(n_components=2)
principal_components_norm = pca_normalized.fit_transform(df_normalized)

principal_components_norm
```

```python
# Plotting the results
plt.figure(figsize=(14, 6))

# without norma
plt.subplot(1, 2, 1)
plt.scatter(principal_components_no_norm[:, 0], principal_components_no_norm[:, 1])
plt.title('PCA without Normalization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

```

```python
# with norm
plt.subplot(1, 2, 2)
plt.scatter(principal_components_norm[:, 0], principal_components_norm[:, 1])
plt.title('PCA with Normalization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.show()

```

```python
# Explained variance
print("Explained variance ratio without normalization:", pca.explained_variance_ratio_)
print("Explained variance ratio with normalization:", pca_normalized.explained_variance_ratio_)
```
